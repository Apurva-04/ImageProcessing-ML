```python
import numpy as np
import keras
import matplotlib.pyplot as plt
```

```python
import keras.preprocessing.image as img_preprocess

img=img_preprocess.load_img('../dataSets/pic2.jpeg',target_size=(128,128))
img_arr=(img_preprocess.img_to_array(img)/255.).reshape(1,128,128,3)
img
#plt.imshow(img_arr)
```

```python
img_preprocess.array_to_img(np.array(keras.layers.UpSampling2D(size=(3,3))(img_arr)).reshape(3*128,3*128,3))
```
