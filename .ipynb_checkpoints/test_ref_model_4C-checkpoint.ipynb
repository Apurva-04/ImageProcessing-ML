```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
h,w=160,160

x_train=[]
y_train=[]
x_test=[]
y_test=[]

path_train_low='../dataSets/ref/our485/low/'
path_train_high='../dataSets/ref/our485/high/'
path_test_low='../dataSets/ref/eval15/low/'
path_test_high='../dataSets/ref/eval15/high/'

print('\nLoding Train dataset :')
i=0;
for img in os.listdir(path_train_low):
    img_low_path=os.path.join(path_train_low, img)
    img_high_path=os.path.join(path_train_high, img)
    
    img_low=load_img(img_low_path, target_size=(h,w))
    img_low=img_to_array(img_low)/255.
       
    i=i+1
    print("{:.2f}".format(i/4.89)+' %',end='\r')
    
    img_high=load_img(img_high_path, target_size=(h,w))
    img_high=img_to_array(img_high)/255.
    
    x_train.append(img_low)
    y_train.append(img_high)
    
print('\nLoding Test dataset :')
i=0;
for img in os.listdir(path_test_low):
    img_low_path=os.path.join(path_test_low, img)
    img_high_path=os.path.join(path_test_high, img)
    
    img_low=load_img(img_low_path, target_size=(h,w))
    img_low=img_to_array(img_low)/255.
   
    i=i+1
    print("{:.2f}".format(i/0.14)+' %',end='\r')
    
    img_high=load_img(img_high_path, target_size=(h,w))
    img_high=img_to_array(img_high)/255.
    
    x_test.append(img_low)
    y_test.append(img_high)

x_train=np.array(x_train, dtype="float32")
y_train=np.array(y_train, dtype="float32")

x_test=np.array(x_test, dtype="float32")
y_test=np.array(y_test, dtype="float32")
```

```python
fig=plt.figure()

f=fig.add_subplot(1,2,1)
s=fig.add_subplot(1,2,2)

f.imshow(x_test[1])
s.imshow(y_test[1])
```

```python
from keras.models import Model
from keras.models import Sequential
from keras.layers import Input,add,Conv2D,MaxPooling2D,UpSampling2D,Flatten,Dropout,Dense,BatchNormalization,Reshape,UpSampling2D,Conv2DTranspose,Concatenate

def InstantiateModel(in_):
    #denoise
    conv0 = Conv2D(64, (3,3), strides=1, padding='same', activation=None)(in_)#(in_)
    
    conv1 = Conv2D(64, (3,3), strides=2, padding='same', activation='relu')(conv0)
    conv2 = Conv2D(64, (2,2), strides=2, padding='same', activation='relu')(conv1)
    conv3 = Conv2D(64, (2,2), strides=2, padding='same', activation='relu')(conv2)
    
    up3 = UpSampling2D((2,2))(conv3)
    deconv3 = Conv2D(64, (2,2), strides=1, padding='same', activation='relu')(up3)+conv2
    up2 = UpSampling2D((2,2))(deconv3)
    deconv2 = Conv2D(64, (3,3), strides=1, padding='same', activation='relu')(up2)+conv1
    up1 = UpSampling2D((2,2))(deconv2)
    deconv1 = Conv2D(64, (3,3), strides=1, padding='same', activation='relu')(up1)+conv0
    
    deconv2_resize = UpSampling2D((2,2))(deconv2)
    feature_gather = Concatenate()([deconv2_resize, deconv1])
    model_1 = Conv2D(3, (1,1), padding='same', activation=None)(feature_gather)
    
    #adjustment
    model_2 = Conv2D(64, (3,3), activation='relu', padding='same', strides=1)(in_)
    model_2 = Conv2D(64, (2,2), activation='relu', padding='same', strides=1)(model_2)
    model_2 = Conv2D(64, (2,2), activation='relu', padding='same', strides=1)(model_2)
    model_2 = Conv2D(3, (1,1), padding='same', activation=None)(model_2)
    
    model_add = Concatenate()([model_2,model_1])
    
    #restoration
    model_3 = Conv2D(64, (2,2), strides=1, padding='same', activation='relu')(model_add)
    model_3 = Conv2D(64, (2,2), activation='relu', padding='same', strides=1)(model_3)
#     model_3 = Conv2D(3,(1,1), activation='sigmoid',padding='same',strides=1)(model_3)
    
#     #adjustment-pro
#     model_4 = Conv2D(32, (2,2), strides=1, padding='same', activation='relu')(model_2)
#     model_4 = Conv2D(32, (2,2), activation='relu', padding='same', strides=1)(model_4)
#     model_4 = Conv2D(1,(1,1), activation=None,padding='same',strides=1)(model_4)
    
#     feature_gather_1 = Concatenate()([model_3, model_4])
#     feature_fusion = Conv2D(64, (1,1), activation='relu', padding='same', strides=1)(feature_gather_1)
    
    out = Conv2D(3, (1,1), activation='sigmoid', padding='same', strides=1)(model_3)#(feature_fusion)
    
    return out
```

```python
Input_Sample = Input(shape=(h, w, 3))
Output_ = InstantiateModel(Input_Sample)
model = Model(inputs=Input_Sample, outputs=Output_)
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model,to_file='model.png',show_shapes=True, show_layer_names=True)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

checkpoint=ModelCheckpoint('models/ref/ref_cnn4.h5',
                            monitor='val_loss',
                            mode='min',
                            save_best_only=True,
                            verbose=1
                          )
earlystop=EarlyStopping(monitor='val_loss',
                        patience=6,
                        min_delta=0,
                        restore_best_weights=True,
                        verbose=1
                       )
reduce_lr=ReduceLROnPlateau(monitor='val_loss',
                            factor=0.2,
                            patience=6,
                            min_delta=0.001,
                            verbose=1
                           )
callbacks=[checkpoint,earlystop,reduce_lr]

model.compile(loss=keras.losses.MeanSquaredError(),
              optimizer=keras.optimizers.Adam(learning_rate=0.001),
              metrics=['accuracy']
             )

history=model.fit(
    x=x_train,
    y=y_train,
    batch_size=32,
    shuffle=True,
    epochs=100,
    validation_data=(x_test, y_test),
    callbacks=callbacks
    )
```

```python
plt.grid()

plt.plot(history.history['loss'],'--^r',label="loss")
plt.plot(history.history['accuracy'],'--^g',label='accuracy')
plt.plot(history.history['val_loss'],'-or',label='val_loss')
plt.plot(history.history['val_accuracy'],'-og',label='val_accuracy')
plt.legend()
```
